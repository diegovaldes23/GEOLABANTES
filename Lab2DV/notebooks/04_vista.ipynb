```python
# %%writefile app.py
import streamlit as st
import pandas as pd
import rasterio
import os
import numpy as np
import leafmap.foliumap as leafmap

%run app.py


# ConfiguraciÃ³n
st.set_page_config(page_title="Pudahuel Dashboard", layout="wide")
st.title("ðŸ›°ï¸ Dashboard de Cambio Urbano - Pudahuel")

# Rutas (AsegÃºrate de que estas carpetas existan)
DATA_DIR = "data/processed"

# Sidebar
analisis = st.sidebar.selectbox("Selecciona Ãndice", ["NDVI (VegetaciÃ³n)", "NDBI (Urbano)"])

# Mapa
m = leafmap.Map(center=[-33.43, -70.78], zoom=12)

if analisis == "NDVI (VegetaciÃ³n)":
    path = os.path.join(DATA_DIR, "delta_ndvi_2017_2024.tif")
    m.add_raster(path, cmap="RdYlGn", layer_name="Delta NDVI")
else:
    path = os.path.join(DATA_DIR, "delta_ndbi_2017_2024.tif")
    m.add_raster(path, cmap="OrRd", layer_name="Delta NDBI")

m.to_streamlit(height=600)

# EstadÃ­sticas
csv_path = "outputs/analisis_zonal_pudahuel.csv"
if os.path.exists(csv_path):
    df = pd.read_csv(csv_path)
    st.write("### Resumen EstadÃ­stico por Cuadrante", df.describe())
```
